## Introduction
In this project, we will be looking at data from US School Improvement Grants in 2010. This program gave nearly $4B to schools to help them renovate or improve their programs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset

```python
df = pd.read_csv('schoolimprovement2010grants.csv')
print(df.head())
```

```python
sns.set_palette("rocket")
```

```python
# Create a displot of the Award Amount
sns.displot(df['Award_Amount'],
             kind='kde',
             rug=True,
             fill=True)

# Plot the results
plt.show()
```

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue='Region')

plt.show()
plt.clf()
```
